```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv ('../processedData/processedDataPilot2_decimal.csv')
print(df)
```

```python
imgNames = np.unique(df['imgName'])
subNames = np.unique(df['subId'])
df['valueDiff']=df['valueDiff'].apply(np.abs)
imgNames
```

```python
df[df['rt'] > 10].mean(axis=0).rt
```

```python
lowDf = df[df['valueDiff']<0.55]
highDf = df[df['valueDiff']>0.55]
lowDf
```

```python
def rtCongruency(df,congr):
    meanrt = df[df['congruency'] ==congr].mean(axis=0).rt
    sdrt = df[df['congruency'] ==congr].std(axis=0).rt
    return meanrt,sdrt
```

```python
lowValueCongruentRt = rtCongruency(lowDf,1)
lowValueInCongruentRt = rtCongruency(lowDf,0)
highValueCongruentRt = rtCongruency(highDf,1)
highValueInCongruentRt = rtCongruency(highDf,0)
```

```python
df["meanPersonRt"]=df["rt"]
df
```

```python
for subId in subNames:
    meanrt =df[df["subId"]==subId].mean(axis=0).rt
    df["meanPersonRt"][df["subId"]==subId] = meanrt
```

```python
df['normRt'] = df['rt']/df['meanPersonRt']
```
